```python
# Questão 2: Validando um endereço de e-mail
```

```python

email = input("Digite um endereço de e-mail: ")
# Verificando se tem um @ e se não é o primeiro ou último caractere
if '@' in email and email.index('@') > 0 and email.index('@') < len(email) - 1:
    # Verificando se termina com .com e tem no mínimo 10 caracteres
    if email.endswith('.com') and len(email) >= 10:
        print("O endereço de e-mail é válido.")
    else:
        print("O endereço de e-mail não atende aos requisitos.")
else:
    print("O endereço de e-mail não atende aos requisitos.")
```
